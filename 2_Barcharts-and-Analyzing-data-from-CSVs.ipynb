if we see our x-axis that is **Ages** changed to indexes, let's fix this.

```python
plt.bar(x_indexes - width, dev_y, width = width, label ='All devs')
plt.bar(x_indexes , py_dev_y, width = width, label = 'Python')
plt.bar(x_indexes + width, js_dev_y,width = width, label = 'JavaScript')


plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.title('Median Salary USD by Age')

plt.legend()

plt.xticks(ticks = x_indexes,labels = ages_x)

plt.show()
```

Parameters:
ticks: array-like containing the list of xtick locations. Passing an empty list removes all xticks. (Optional)
labels: array-like containing the labels to place at the given ticks locations. This argument can only be passed if ticks are passed as well. (Optional)

Return Type
The Matplotlib xticks() function returns:
locs: The list of xtick locations.
labels: The list of xlabel text objects.

```python
import csv
import numpy as np
from collections import Counter
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
with open ('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
    
print(language_counter)
```

```python
#print(language_counter.most_common(15))
most_common_language = []
most_common_language = language_counter.most_common(15) #most_common is a method
print(most_common_language)
```

```python
#let's plot the above data
```

```python
languages = []
popularity = []
```

```python
for item in most_common_language:
    languages.append(item[0])
    popularity.append(item[1])
    
print(languages)
print('\n')
print(popularity)
```

```python
plt.style.use('fivethirtyeight')
plt.barh(languages,popularity)
plt.title('Most Popular Languages')
plt.xlabel('Programming Languages')
plt.ylabel('No of People with use')

plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.barh(languages,popularity)
plt.title('Most Popular Languages')
#plt.ylabel('Programming Languages')
plt.xlabel('No of People with use')

plt.show()
```

```python
languages.reverse()
popularity.reverse()
```

```python
plt.style.use('fivethirtyeight')
plt.barh(languages,popularity)
plt.title('Most Popular Language')
plt.xlabel('No of People with use')

plt.show()
```
