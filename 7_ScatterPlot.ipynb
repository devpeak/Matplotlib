```python
data = pd.read_csv('2019-05-31-data.csv')
```

```python
data.columns
```

```python
data.tail(5)
```

```python
data.head()
```

```python
views = data['view_count']
likes = data['likes']
ratio = data['ratio']

print(ratio)
```

```python
plt.scatter(views, likes, edgecolor = 'cyan', linewidth = 1, alpha = 0.75)

plt.xscale('log')
plt.yscale('log')
plt.title('Popular Youtube Videos')
plt.xlabel('Total Views')
plt.ylabel('Total Likes')
plt.show()
```

```python
plt.scatter(views, likes, c = ratio, cmap = 'summer',
            edgecolor = 'black', linewidth = 1, alpha = 0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Most Popular Videoss')
plt.xlabel('Total Views')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.savefig('scatterplot.png',bbox_inches = 'tight')
plt.show()
```

```python
#vertical bar on the right side of plot is colorbar
#dot which are bluish-green are performed less which are around 50-70 i.e
#they have the same number of likes as well dislike

#whereas, bright yello performed better [in 90s]
```
