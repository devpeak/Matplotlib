```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
data = pd.read_csv('Dev.csv')
```

```python
data.columns
```

```python
ages = data['Age']
```

```python
py_salaries = data['Python']
js_salaries = data['JavaScript']
dev_salaries = data['All_Devs']
```

```python
plt.plot(ages,py_salaries,label = 'Python')
plt.plot(ages,js_salaries,label = 'JavaScript')
plt.plot(ages, dev_salaries, color = 'red', linestyle = '--', label = 'All_Devs')

plt.legend()
plt.title('Median Salary (USD) By Age')
plt.xlabel('Age')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

We created two variables, **fig** and **ax**. Remember, these are arbitrary names but a standard and we are one of the good guys, so we will follow the convention.

These two variables now hold the two core objects used for all types of plotting operations. First object fig, short for figure, imagine it as the **frame** of your plot. You can **resize, reshape the frame but you cannot draw on it**. On a single notebook or a script, you can have multiple figures. Each figure can have multiple subplots

Here, subplot is synonymous with axes. The second object, ax, short for axes, is the canvas you **draw on**. Or rephrasing, it is the blank sheet you can **plot and hold your data**. An axes object c**an only belong to one figure**.

Axes methods vs. pyplot, understanding further
At the beginning of the post, I said that pyplot was a more beginner-friendly method to interact with Matplotlib. It is true that compared to axes methods, pyplot offers a quicker and more concise method of plotting. It will have less local variables and syntax. But why most people prefer the object-oriented way?

pyplot **cannot create new axes or a new figure** and intelligently plot the new data. As we get to more complex plotting like this one, we are going to need a more flexible approach.

replace all plt with ax

```python
fig, ax = plt.subplots()
ax.plot(ages,py_salaries,label  = 'Python')
ax.plot(ages,js_salaries,label = 'JavaScript')
ax.plot(ages,dev_salaries, label = 'All_Devs')

ax.legend()
ax.set(title = 'Median Salary(USD) by Age',xlabel = 'Ages',ylabel = 'Median Salary (USD)');

plt.tight_layout()
plt.show()
```

```python
fig , (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1)

ax1.plot(ages,py_salaries,label  = 'Python',color = 'red', linewidth = 3)
ax1.plot(ages,dev_salaries, label = 'All_Devs', color = 'green' ,
         linewidth = 3 , linestyle = '--')

ax2.plot(ages,js_salaries,label = 'JavaScript', color = 'blue',linewidth = 3)
ax2.plot(ages,dev_salaries, label = 'All_Devs', color = 'yellow' ,
         linewidth = 3 , linestyle = '--')

ax1.legend()
ax2.legend()

ax1.set(title = 'Median Salary(USD) by Age',xlabel = 'Ages',ylabel = 'Median Salary (USD)');
ax2.set(title = 'Median Salary(USD) by Age',xlabel = 'Ages',ylabel = 'Median Salary (USD)');

plt.savefig('subplots.png')
plt.tight_layout()
plt.show()
```
