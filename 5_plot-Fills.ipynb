# FILLING AREA ON LINE PLOTS

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data_fills.csv')
```

```python
ages = data['Age']
```

```python
dev_salaries = data['All_Devs']
```

```python
py_salaries = data['Python']
```

```python
js_salaries = data['JavaScript']
```

```python
plt.plot(ages,dev_salaries,color = '#B8255F', label = 'All Devs', linestyle = '-', linewidth = 3)
plt.plot(ages,py_salaries,color = '#d95f02',label = 'Python Devs', linestyle = '-')
plt.plot(ages,js_salaries,color = '#0f0f0f',label = 'JavaScript Devs',linestyle = ':',linewidth = 3)


plt.title('Median Salaries (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
#adding fields
```

```python
plt.plot(ages,dev_salaries,color = '#B8255F', label = 'All Devs', linestyle = '-', linewidth = 3)
plt.plot(ages,py_salaries,color = '#d95f02',label = 'Python Devs', linestyle = '-')
plt.plot(ages,js_salaries,color = '#0f0f0f',label = 'JavaScript Devs',linestyle = ':',linewidth = 3)


plt.title('Median Salaries (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.fill_between(ages,py_salaries)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
adding alpha
```

```python
plt.plot(ages,dev_salaries,color = '#B8255F', label = 'All Devs', linestyle = '-' )
plt.plot(ages,py_salaries,color = '#d95f02',label = 'Python Devs', linestyle = '-')
plt.plot(ages,js_salaries,color = '#0f0f0f',label = 'JavaScript Devs',linestyle = ':')


plt.title('Median Salaries (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.fill_between(ages,py_salaries,alpha = 0.50)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
overall_salary = 57287
```

```python
plt.plot(ages,dev_salaries,color = '#B8255F', label = 'All Devs', linestyle = '-' )
plt.plot(ages,py_salaries,color = '#d95f02',label = 'Python Devs', linestyle = '-')
plt.plot(ages,js_salaries,color = '#0f0f0f',label = 'JavaScript Devs',linestyle = ':')


plt.title('Median Salaries (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.fill_between(ages,py_salaries,overall_salary,alpha = 0.30)
plt.legend()
plt.tight_layout()
plt.show()
```

The above figures states that when the python salary is lesser than
57287, it is plots above the line, but when it surpases the **overall salary** it plots below the line.

```python
#we can also fill some additional condition
```

```python
plt.plot(ages,dev_salaries,color = '#B8255F', label = 'All Devs', linestyle = '-' )
plt.plot(ages,py_salaries,color = '#d95f02',label = 'Python Devs', linestyle = '-')
plt.plot(ages,js_salaries,color = '#0f0f0f',label = 'JavaScript Devs',linestyle = ':')


plt.title('Median Salaries (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.fill_between(ages, py_salaries, overall_salary,
                 where = (py_salaries > overall_salary) ,
                 interpolate = True, color = '#43a2ca', alpha = 0.50)

plt.fill_between(ages, py_salaries, overall_salary,
                 where = (py_salaries < overall_salary) ,
                 interpolate = True , color = '#fc9272', alpha = 0.50)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#interpolate just insures that specific aroea get filled correctly
```

let's plot for a condition b.w. All Devs and Python Devs

```python
plt.plot(ages,dev_salaries,color = '#B8255F', label = 'All Devs', linestyle = '-' )
plt.plot(ages,py_salaries,color = '#d95f02',label = 'Python Devs', linestyle = '-')
plt.plot(ages,js_salaries,color = '#0f0f0f',label = 'JavaScript Devs',linestyle = ':')


plt.title('Median Salaries (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.fill_between(ages, py_salaries, dev_salaries,
                 where = (py_salaries > dev_salaries) ,
                 interpolate = True, color = "#756bb1", alpha = 0.70)

plt.fill_between(ages, py_salaries, dev_salaries,
                 where = (py_salaries <= dev_salaries) ,
                 interpolate = True , color = '#fc9272', alpha = 0.50)


plt.legend()
plt.tight_layout()
plt.show()
```

```python
#above plot is emphasizing that python develpers are almost paid above than 
#normal devs.
```

```python
plt.plot(ages,dev_salaries,color = '#B8255F', label = 'All Devs', linestyle = '-' )
plt.plot(ages,py_salaries,color = '#d95f02',label = 'Python Devs', linestyle = '-')
plt.plot(ages,js_salaries,color = '#0f0f0f',label = 'JavaScript Devs',linestyle = ':')


plt.title('Median Salaries (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.fill_between(ages, py_salaries, dev_salaries,
                 where = (py_salaries > dev_salaries) ,
                 interpolate = True, color = "#756bb1", alpha = 0.70,
                 label = 'Above Avg')

plt.fill_between(ages, py_salaries, dev_salaries,
                 where = (py_salaries <= dev_salaries) ,
                 interpolate = True , color = '#fc9272', alpha = 0.50,
                 label = 'Below Avg')


plt.legend()
plt.tight_layout()
plt.savefig
plt.show()
```
