```python
programming_language = ['JavaScript','Python','HTML/CSS','SQL','JAVA']
popularity = [69.7,41.6,62.4,56.9,38.4]
colors = ['#ffffcc','#a1dab4','#41b6c4','#2c7fb8','#253494']
colors.reverse()
```

```python
plt.title('Most Popular Languages')
plt.pie(popularity, labels = programming_language, colors = colors, 
       wedgeprops = {'edgecolor':'black'})
plt.tight_layout()
plt.savefig('piechart.png')
plt.show()
```

# ADDING EXPLODE TO PIECHART

```python
explode = [0,0.2,0,0,0]
plt.title('Most Popular Languages')

plt.pie(popularity, labels = programming_language, colors = colors,
        explode = explode, wedgeprops = {'edgecolor':'black'})

plt.tight_layout()
plt.savefig('piechart.png')
plt.show()
```

```python
#0.2 states 20% of radius is exploded out here
```

# ADDING SHADOW 

```python
explode = [0,0.2,0,0,0]
plt.title('Most Popular Languages')

plt.pie(popularity, labels = programming_language, colors = colors,
        explode = explode, shadow = True, wedgeprops = {'edgecolor':'black'})

plt.tight_layout()
plt.savefig('piechart.png')
plt.show()
```

# ADDING startangle

The angle by which the start of the pie is rotated, counterclockwise from the x-axis.

```python
explode = [0,0.2,0,0,0]
plt.title('Most Popular Languages')

plt.pie(popularity, labels = programming_language, colors = colors,
        explode = explode, shadow = True, startangle = 90,
        wedgeprops = {'edgecolor':'black'})

plt.tight_layout()
plt.savefig('piechart.png')
plt.show()
```

# SHOWING PERCENTAGE IN PIECHART

```python
explode = [0,0.2,0,0,0]
plt.title('Most Popular Languages')

plt.pie(popularity, labels = programming_language, colors = colors,
        explode = explode, shadow = True, startangle = 90,
        autopct = '%1.1f%%', wedgeprops = {'edgecolor':'black'})

plt.tight_layout()
plt.savefig('piechart.png')
plt.show()
```
